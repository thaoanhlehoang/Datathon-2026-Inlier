```python
import pandas as pd
import duckdb
con = duckdb.connect()
```

```python
query = """
    SELECT g.region, SUM(oi.quantity * oi.unit_price) AS total_revenue
    FROM '../../data/master/geography.csv' g
    JOIN '../../data/transaction/orders.csv' o ON g.zip = o.zip
    JOIN '../../data/transaction/order_items.csv' oi
    ON o.order_id = oi.order_id
    GROUP BY g.region
    ORDER BY total_revenue DESC
"""
```

```python
df = con.execute(query).df()
df
```
