```python
import pandas as pd
import duckdb
import numpy as np

con = duckdb.connect()
```

```python
query = """
    SELECT r.return_reason, COUNT(r.return_reason) AS count_reason
    FROM '../../data/transaction/returns.csv' r
    JOIN '../../data/master/products.csv' p ON r.product_id = p.product_id
    WHERE p.category = 'Streetwear'
    GROUP BY r.return_reason
    ORDER BY count_reason DESC
"""
```

```python
df = con.execute(query).df()
df
```
