```python
import pandas as pd
import duckdb
con = duckdb.connect()
```

```python
query = """
    SELECT traffic_source, AVG(bounce_rate) as avg_bounce_rate
    FROM '../../data/operational/web_traffic.csv'
    GROUP BY traffic_source
    ORDER BY avg_bounce_rate
"""
```

```python
df = con.execute(query).df()
df
```
