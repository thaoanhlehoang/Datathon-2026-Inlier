```python
import pandas as pd
import duckdb
import numpy as np

con = duckdb.connect()
```

```python
query = """
    SELECT
        segment, 
        AVG(1 - cogs/price) AS avg_profit
    FROM '../../data/master/products.csv'
    GROUP BY segment
    ORDER BY avg_profit DESC
    LIMIT 1
"""
```

```python
df = con.execute(query).df()
df
```
