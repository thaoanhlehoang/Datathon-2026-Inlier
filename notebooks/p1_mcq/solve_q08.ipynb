```python
import pandas as pd
import duckdb
con = duckdb.connect()
```

```python
query = """
    SELECT payment_method, COUNT(payment_method) as count
    FROM '../../data/transaction/orders.csv'
    WHERE order_status = 'cancelled'
    GROUP BY payment_method
    ORDER BY count DESC
"""
```

```python
df = con.execute(query).df()
df
```
