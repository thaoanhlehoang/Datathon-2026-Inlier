```python
import pandas as pd
import duckdb
con = duckdb.connect()
```

```python
query = """
    SELECT installments, AVG(payment_value) as avg_value
    FROM '../../data/transaction/payments.csv'
    GROUP BY installments
    ORDER BY avg_value DESC
"""
```

```python
df = con.execute(query).df()
```
