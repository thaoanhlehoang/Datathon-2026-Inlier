```python
import pandas as pd
import duckdb
con = duckdb.connect()
```

```python
query = """
    WITH order_count AS (
        SELECT p.size, COUNT(*) AS total_order
        FROM '../../data/transaction/order_items.csv' oi
        JOIN '../../data/master/products.csv' p
        ON oi.product_id = p.product_id
        GROUP BY p.size
    ),
    return_count AS (
        SELECT p.size, COUNT(*) as total_return
        FROM '../../data/transaction/returns.csv' r
        JOIN '../../data/master/products.csv' p
        ON r.product_id = p.product_id
        GROUP BY p.size
    )

    SELECT o.size, (total_return * 1.0 / total_order) as return_rate
    FROM order_count o
    JOIN return_count r ON o.size = r.size
    ORDER BY return_rate DESC
"""
```

```python
df = con.execute(query).df()
```
